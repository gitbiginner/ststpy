```python
%%time
from sklearn.feature_selection import RFE
from sklearn.ensemble import GradientBoostingRegressor

IDTW60 = []

i=0
while i < 27:
    dropped = []
    listy = list(pnxrt['y'][i:i+121])
    dfcs = listdfs[i]
    dfcs['y']= listy

    train = dfcs[0:120]
    valid = dfcs[120:121]
    feat_cols = train.drop(columns=["Unnamed: 0", "y"]).columns.tolist()

    X_train = train[feat_cols]
    X_valid = valid[feat_cols]
    y_train = train["y"]
    y_valid = valid['y']

    # estimator として GradientBoostingRegressor を使用。特徴量を60個選択
    selector = RFE(GradientBoostingRegressor(n_estimators=100, random_state=42), n_features_to_select=60)
    selector.fit(X_train, y_train)
    mask = selector.get_support()
    num = len(X_train.columns)
    #num = str(num)

    for n in range(num):
        if    mask[n]==False:
            dropped.append(X_train.columns[n])

    feat_cols = train.drop(columns=["Unnamed: 0", "y"]+dropped).columns.tolist()

    X_train = train[feat_cols]
    X_valid = valid[feat_cols]
    y_train = train["y"]
    y_valid = valid['y']

    import xgboost as xgb
    dtrain = xgb.DMatrix(X_train, y_train)
    dvalid = xgb.DMatrix(X_valid, y_valid)

    params = {
        "objective" : "reg:squarederror",
        "eval_metric" : "rmse"
    }

    model = xgb.train(
        params = params,
        dtrain = dtrain,
        evals = [(dtrain, "train"), (dvalid, "valid")],
        num_boost_round = 100
    )

    pred = model.predict(xgb.DMatrix(X_valid))
    IDTW60.append(pred)
    i += 1

print('end')

```

```python
%%time

from sklearn.feature_selection import RFE
from sklearn.ensemble import GradientBoostingRegressor

predlistcs60= []


i=0
while i < 27:
    dropped = []
    listy = list(pnxrt['y'][i:i+121])
    dfcs = listdfcs[i]
    dfcs['y']= listy

    train = dfcs[0:120]
    valid = dfcs[120:121]
    feat_cols = train.drop(columns=["Unnamed: 0", "y"]).columns.tolist()

    X_train = train[feat_cols]
    X_valid = valid[feat_cols]
    y_train = train["y"]
    y_valid = valid['y']

    # estimator として GradientBoostingRegressor を使用。特徴量を60個選択
    selector = RFE(GradientBoostingRegressor(n_estimators=100, random_state=42), n_features_to_select=60)
    selector.fit(X_train, y_train)
    mask = selector.get_support()
    num = len(X_train.columns)
    #num = str(num)

    for n in range(num):
        if    mask[n]==False:
            dropped.append(X_train.columns[n])

    feat_cols = train.drop(columns=["Unnamed: 0", "y"]+dropped).columns.tolist()

    X_train = train[feat_cols]
    X_valid = valid[feat_cols]
    y_train = train["y"]
    y_valid = valid['y']

    import xgboost as xgb
    dtrain = xgb.DMatrix(X_train, y_train)
    dvalid = xgb.DMatrix(X_valid, y_valid)

    params = {
        "objective" : "reg:squarederror",
        "eval_metric" : "rmse"
    }

    model = xgb.train(
        params = params,
        dtrain = dtrain,
        evals = [(dtrain, "train"), (dvalid, "valid")],
        num_boost_round = 100
    )

    pred = model.predict(xgb.DMatrix(X_valid))
    predlistcs60.append(pred)
    i += 1

print('end')

```
