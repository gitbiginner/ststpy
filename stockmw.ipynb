```python
#make next month return
mname = []
mnum = []
stmonth = []
nxrt1a = []
nxrt  =  []

# from 0 to 156   #200903-202204
for i in range(1,len(madedf.columns)-1):
  month = madedf.columns[i] + ">" + madedf.columns[i+1]
  startmonth = madedf.columns[i] 
  premonthdays = len(madedf[madedf.columns[i]].dropna())-1
  postmonthdays = len(madedf[madedf.columns[i+1]].dropna())-1
  prelastprice = madedf[madedf.columns[i]][premonthdays]
  postlastprice = madedf[madedf.columns[i+1]][postmonthdays]
  retans = (postlastprice/prelastprice)-1



  mname.append(month)
  mnum.append(i)
  stmonth.append(startmonth)
  nxrt.append(retans)
print('end')
```

```python
nxrt1a = []
for i in range(0, len(madedf.columns)-2):
    premonthdays1a = len(madedf[madedf.columns[i]].dropna())-1
    postmonthdays1a = len(madedf[madedf.columns[i+1]].dropna())-1
    prelastprice1a = madedf[madedf.columns[i]][premonthdays1a]
    postlastprice1a = madedf[madedf.columns[i+1]][postmonthdays1a]
    retans1a = (postlastprice1a/prelastprice1a)-1
    nxrt1a.append(retans1a)
print("D")
```

```python
import pandas as pd
import numpy as np

from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
from tqdm import tqdm

monStock = madedf.drop(columns=["2009-02"]) ##change from dfname

listm = monStock.columns


starti = 12

premonthcount = len(madedf.columns)-1

pbar = tqdm(premonthcount)
dflist = []
while starti < premonthcount-1:

    idtwcalc = []
    for p  in range(starti-12, starti+1):
        datap = monStock[monStock.columns[p]]
        datap = datap/monStock[monStock.columns[p]][0]
        datap = datap.dropna()
        t = monStock[monStock.columns[starti]]
        t = t/monStock[monStock.columns[starti]][0]
        t = t.dropna()

        dtw_distance, warp_path = fastdtw(datap, t, dist=euclidean)
        idtw = dtw_distance

        idtwcalc.append(idtw)


    idtwcalc = np.array(idtwcalc)
    idtwcalcres0  = idtwcalc.reshape(1,13) # decline var


    exec("df"+str(stmon[starti]) +"= pd.DataFrame(data = idtwcalcres0, index=[monStock.columns[starti]], columns = mnths)")
    exec("dfname" +"= df"+str(stmon[starti]))
    dflist.append(dfname)
  



    starti += 1
    pbar.update(1)
pbar.close()

print('end')
```

```python
predlist = []

df = dfedit

for i in range(len(dfedit)-1):

    train = df[i][1:12]
    valid = df[i][0:1]
    
    #feat_cols = train.drop("date", "mname","nxrt")
    
    feat_cols = train.drop(columns=["date", "mname", "nxrt"]).columns.tolist()

    X_train = train[feat_cols]
    X_valid = valid[feat_cols]
    y_train = train["nxrt"]
    y_valid = valid['nxrt']

    import xgboost as xgb
    dtrain = xgb.DMatrix(X_train, y_train)
    dvalid = xgb.DMatrix(X_valid, y_valid)

    params = {
        "objective" : "reg:squarederror",
        "eval_metric" : "rmse"
    }

    model = xgb.train(
        params = params,
        dtrain = dtrain,
        evals = [(dtrain, "train"), (dvalid, "valid")],
        num_boost_round = 25
    )

    pred = model.predict(xgb.DMatrix(X_valid))
    predlist.append(pred)

print('end')
```
