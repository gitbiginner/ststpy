```python
#make next month return
mname = []
mnum = []
stmonth = []
nxrt  =  []

# from 0 to 156   #200903-202204
for i in range(len(madedf.columns)-1):
  month = madedf.columns[i] + ">" + madedf.columns[i+1]
  startmonth = madedf.columns[i] 
  premonthdays = len(madedf[madedf.columns[i]].dropna())-1
  postmonthdays = len(madedf[madedf.columns[i+1]].dropna())-1
  prelastprice = madedf[madedf.columns[i]][premonthdays]
  postlastprice = madedf[madedf.columns[i+1]][postmonthdays]
  retans = (postlastprice/prelastprice)-1

  mname.append(month)
  mnum.append(i)
  stmonth.append(startmonth)
  nxrt.append(retans)
print('end')
```

```python
import pandas as pd


returns = pd.DataFrame()
returns["mname"] = mname
returns["mnum"] = mnum
returns["nxrt"] = nxrt
returns
```

```python
import pandas as pd
mmnum  = pd.DataFrame()
mmnum["date"] = stmonth
mmnum["mnum"] = mnum
mmnum
```

```python
import pandas as pd
import numpy as np

from scipy.spatial.distance import euclidean
from fastdtw import fastdtw
from tqdm import tqdm

monStock = madedf ##change from dfname

listm = monStock.columns


starti = 12

premonthcount = len(madedf.columns)

pbar = tqdm(premonthcount)
dflist = []
while starti < premonthcount-1:

    idtwcalc = []
    for p  in range(starti-12, starti+1):
        datap = monStock[monStock.columns[p]]
        datap = datap/monStock[monStock.columns[p]][0]
        datap = datap.dropna()
        t = monStock[monStock.columns[starti]]
        t = t/monStock[monStock.columns[starti]][0]
        t = t.dropna()

        dtw_distance, warp_path = fastdtw(datap, t, dist=euclidean)
        idtw = dtw_distance

        idtwcalc.append(idtw)


    idtwcalc = np.array(idtwcalc)
    idtwcalcres0  = idtwcalc.reshape(1,13) # decline var


    exec("df"+str(stmon[starti]) +"= pd.DataFrame(data = idtwcalcres0, index=[monStock.columns[starti]], columns = [listm[starti-12:starti+1]])")
    exec("dfname" +"= df"+str(stmon[starti]))
    dflist.append(dfname)
  



    starti += 1
    pbar.update(1)
pbar.close()

print('end')
```
